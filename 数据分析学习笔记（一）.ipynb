```python
from matplotlib import pyplot as plt
import matplotlib
from matplotlib import font_manager
```

```python
x = range(2, 25, 2)
y = [20,22,22,23,24,23,23,24,25,20,21,18]
plt.plot(x, y)
plt.show()  # 显示图片
```

### 3.调整x,y轴刻度间的距离

```python
plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y)

plt.xticks(x) 
# plt.xticks(x)  设置x为刻度
# plt.xticks(x[::2])
plt.yticks(range(min(y), max(y)+1))
# * 需要控制刻度的稀疏时，采用列表取步长的方法来解决
plt.show()
```

### 4.添加x,y轴的描述信息

```python
# 关于显示中文的方法：
# 第一种 （不知道familly字体名称时就不会显示）
font = {'family': 'youyuan',
        'weight': 'bold',}
matplotlib.rc('font', **font)

# 第二种 (一定可以正常显示)
# my_font = font_manager.FontProperties(fname="C:\Windows\Fonts\msyh.ttc")
# 用第二种方法就哪里要用就添加 fontproperties=font 这个参数
plt.xlabel('时间')
plt.ylabel('速度')

plt.show()
```

### 5.设置图表的网格

```python
plt.plot(x,y)

xticks_label = ['{}点'.format(i) for i in x]
plt.xticks(x, xticks_label) 
plt.yticks(range(min(y),30))
# 让折线图上方留更多空间只需要加大轴上最大的值

# 要显示中文其实就是定义个变量来覆盖x原先的数字值

plt.grid(alpha=0.4)  # alpha设置网格不透明度

# 网格线条对应x,y轴的个数（改变x,y轴的数量可改变网格的密集和稀疏程度）

plt.show()
```

### 6.线条的样式

```python
plt.plot(x, y, color='red',linestyle='-.') 
plt.xticks(x) 
plt.yticks(range(min(y), max(y)+1))

plt.show()
```

### 7.图表图例的设置

+ 第一步：给plt.plot( )中参数label=''传入字符串类型的值，也就是图例的名称

+ 第二步：使用plt.legend( )使上述代码产生效果

```python
plt.plot(x, y, label='随便啥', color='red',linestyle='-')
# label设置图例要显示的文字
plt.xticks(x) 
plt.yticks(range(min(y), max(y)+1))

plt.legend()
# 因为前面设置的中文显示是全局可用的，所以不需要单独在设置
# 如果是用第二种中文显示方法 则通过 prop=font 来显示， 只有图例显示中文是prop,其他依然是fontproperties来设置

plt.show()
```

### 完整的图表

```python
from random import choice
from matplotlib import pyplot as plt, font_manager
# 案例
# 假设在30岁时，根据自己实际情况，统计出来了从11岁到30岁每年交男女朋友数量列表a，
# 请绘制出该数据的折线图，以便分析每年交友数量的走势
# x轴表示年龄， y轴表示数量

a = range(7)
# 随机数量
random_num = [choice(a) for _ in range(20)]
# 年龄
age = range(11, 31)
# 设置图片大小
plt.figure(figsize=(20, 8), dpi=80)
# 绘制折线图  （label设置图例内容）
plt.plot(age, random_num, label='小王', color='red')
# 显示图例
plt.legend()
# 中文字体设置
my_font = font_manager.FontProperties(fname="C:\Windows\Fonts\msyh.ttc")

x_labels = [f'{i}岁' for i in age]
# 这里需要两个参数，一个数据本身，一个字符串形式 用字符串数据去替换数字的数据
plt.xticks(age,  x_labels, fontproperties=my_font)
# 设置y轴刻度
plt.yticks(range(0, 11))
# 设置图标描述信息
plt.xlabel('年龄',  fontproperties=my_font)
plt.ylabel('数量', fontproperties=my_font)

plt.title('不同年龄段的交友折线图',  fontproperties=my_font)

# 绘制网格, plt.grid() 可以传一个参数，也可以默认
plt.grid(alpha=0.6)  # alpha 透明度
# 需要调整网格，只需要改变x、y轴的数量

# 保存图片
plt.savefig('./test3.png')

plt.show()
```
