## anaconda的使用

### 1. 新建conda的虚拟环境
    1.conda create -n envName python=3.8 或 conda create --name envName python=3.8
    
### 2.激活虚拟环境
    + activate envName 
    
### 3.退出虚拟环境
    + deactivate envName
    
### 4.删除虚拟环境
    + conda remove --name envName --all
    + 通常我回直接取目录下删除

## matplotlib库

### 学习绘制折线图

### 1.导入模块 from matplotlib import pyplot as plt 

### 2. 设置图形大小和分辨率
    plt.figure(figsize=(20, 8), dpi=80)
    
### 3.绘制图形
    plt.plot(x, y)
    + x: 坐标轴x上的所有x的值
    + y: 坐标轴y上的所有y的值

### 4.调整x/y轴的刻度
    + plt.xticks(x)
    + plt.yticks(y)
    1. 如果想调整间距，需要传一个参数（包含数字的可迭代对象[], 设置合适的步长即可）
    2. 如果想在x/y轴显示字符串，需要两个参数（两个可迭代对象[],数字的可迭代对象最终会被字符串的可迭代对象占据）或数字与字符串一一对应后只显示字符串。

### 5.显示图形
    + plt.show()
    
### 6.保存图片（支持svg）
    + plt.savefig("file_path")
    + plt.savefig("./test.jpg")
    
### 7.显示中文
    方法1：
        matplotlib.rc(不推荐)
    方法2：
        导入模块 from matplotlib import font_manager
        实例化  my_font = font_manager.FontProperties(fname="C:\Windows\Fonts\msyh.ttc")
        使用 在需要显示中文的地方添加 fontproperties=my_font
        特殊：在图例中使用是 prop=my_font

### 8.多数据折线图
    + 将 plt.plot() 调用多次即可
    + 添加图例（分辨数据谁是谁）
        1. plt.plot(label="mySelf")
            plt.plot(label="Other")
        
        2. plt.legend()
            可以指定参数有 loc ,指定图例显示位置
            prop  设置中文显示

### 9.图形样式更改（折线图的样式）
    > 在plt.plot(color='red, linestyle='--', linewidth=5)中设置
    + color  # 线条颜色 rgb 16进制编码
    + linestyle  # 线条样式
        - 实线
        -- 虚线
        -. 点划线
        : 点虚线
        '' 空，无线条
    + linewidth  # 线条粗细
    + alpha

### 10.添加图形描述
    + plt.xlabel("描述信息")
    + plt.ylabel("描述信息")
    + plt.title("标题信息")
 
### 11.设置网格
    + plt.grid(alpha=0.4)
    alpha 设置透明度
    同样可以设置颜色，样式，粗细
